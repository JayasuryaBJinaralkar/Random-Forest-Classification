```python
sex=pd.get_dummies(titanic_data["Sex"],drop_first=True)
sex.head(5)
```

```python
embark=pd.get_dummies(titanic_data["Embarked"],drop_first=True)
embark.head(5)
```

```python
pclass=pd.get_dummies(titanic_data["Pclass"],drop_first=True)
pclass.head(5)
```

```python
titanic_data=pd.concat([titanic_data,sex,embark,pclass],axis=1)
```

```python
titanic_data.head(5)
```

```python
#Dropping columes
titanic_data.drop(['Sex','Embarked','Passengerid'],axis=1,inplace=True)
titanic_data.head(5)
```

```python
titanic_data.head()


```

```python
X= titanic_data.drop("2urvived",axis=1)
y= titanic_data["2urvived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier

dtc = RandomForestClassifier()
dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_test)
```

```python
# accuracy_score
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
dtc_train_acc = accuracy_score(y_train, dtc.predict(X_train))
dtc_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy of Random forest is : {dtc_train_acc}")
print(f"Test accuracy of Random forest is : {dtc_test_acc}")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
